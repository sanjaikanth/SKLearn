```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.datasets import make_blobs

# Create a synthetic dataset with clusters and outliers
X, _ = make_blobs(n_samples=300, centers=2, cluster_std=1.0, random_state=42)
X = np.vstack([X, np.random.uniform(low=-6, high=6, size=(20, 2))])  # Adding outliers

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], color='b', s=5, label="Inliers")
plt.title("Dataset with Outliers")
plt.show()

# Isolation Forest Model
iso_forest = IsolationForest(contamination=0.1, random_state=42)
y_pred_if = iso_forest.fit_predict(X)
outliers_if = X[y_pred_if == -1]  # Outliers have label -1

# Local Outlier Factor Model
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred_lof = lof.fit_predict(X)
outliers_lof = X[y_pred_lof == -1]  # Outliers have label -1

# Plot results
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
# Plot Isolation Forest results
axs[0].scatter(X[:, 0], X[:, 1], color='b', s=5, label="Inliers")
axs[0].scatter(outliers_if[:, 0], outliers_if[:, 1], color='r', s=20, label="Outliers (Isolation Forest)")
axs[0].set_title("Outlier Detection with Isolation Forest")
axs[0].legend()

# Plot Local Outlier Factor results
axs[1].scatter(X[:, 0], X[:, 1], color='b', s=5, label="Inliers")
axs[1].scatter(outliers_lof[:, 0], outliers_lof[:, 1], color='r', s=20, label="Outliers (LOF)")
axs[1].set_title("Outlier Detection with Local Outlier Factor")
axs[1].legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

# Create a synthetic dataset with clusters and outliers
X, _ = make_blobs(n_samples=300, centers=2, cluster_std=1.0, random_state=42)
X = np.vstack([X, np.random.uniform(low=-6, high=6, size=(20, 2))])  # Adding outliers

# Isolation Forest Model
iso_forest = IsolationForest(contamination=0.1, random_state=42) #Chage contamination to auto to auto select
y_pred_if = iso_forest.fit_predict(X)
outliers_if = X[y_pred_if == -1]  # Outliers have label -1
inliers_if = X[y_pred_if != -1]   # Inliers have label 1

# Plot results side by side
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot original dataset with outliers
axs[0].scatter(X[:, 0], X[:, 1], color='b', s=5, label="Inliers and Outliers")
axs[0].set_title("Dataset with Outliers")
axs[0].legend()

# Plot Isolation Forest results highlighting detected outliers
axs[1].scatter(inliers_if[:, 0], inliers_if[:, 1], color='b', s=5, label="Inliers")
axs[1].scatter(outliers_if[:, 0], outliers_if[:, 1], color='r', s=20, label="Detected Outliers (Isolation Forest)")
axs[1].set_title("Outliers Detected by Isolation Forest")
axs[1].legend()

plt.tight_layout()
plt.show()
```
