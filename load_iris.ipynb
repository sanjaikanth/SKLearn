```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the iris dataset
iris = load_iris()

# Create a DataFrame to view the data easily
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

iris_df
```

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import pandas as pd

# Load the digits dataset
digits = load_digits()

# Create a DataFrame to view data easily
digits_df = pd.DataFrame(data=digits.data, columns=[f'pixel_{i}' for i in range(digits.data.shape[1])])
digits_df['target'] = digits.target

digits_df
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load the California Housing dataset
california_housing = fetch_california_housing()

# Create a DataFrame to view data easily
housing_df = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)
housing_df['target'] = california_housing.target

housing_df
```

```python
from sklearn.datasets import fetch_lfw_people
import matplotlib.pyplot as plt

# Load the LFW dataset (specify minimum number of images per person to limit dataset size if desired)
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4)

# Extract the features (images) and target labels (person names)
X = lfw_people.data
y = lfw_people.target
target_names = lfw_people.target_names
images = lfw_people.images

# Display the shape of the dataset
print(f"Data shape (flattened images): {X.shape}")
print(f"Image shape (as loaded): {images.shape}")
print(f"Number of unique targets (people): {len(target_names)}")
print(f"Target names:\n{target_names}")
```

```python
# Plot a few sample images with their labels
fig, axes = plt.subplots(2, 5, figsize=(15, 8), subplot_kw={'xticks':[], 'yticks':[]})
for i, ax in enumerate(axes.flatten()):
    ax.imshow(images[i], cmap='gray')
    ax.set_title(target_names[y[i]])
plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import pandas as pd

# Generate a synthetic classification dataset
X, y = make_classification(
    n_samples=1000,     # Total number of samples
    n_features=10,      # Total number of features
    n_informative=5,    # Number of informative features
    n_redundant=2,      # Number of redundant features
    n_classes=2,        # Number of classes (binary classification)
    random_state=42     # Seed for reproducibility
)

# Create a DataFrame to view the data easily
data_df = pd.DataFrame(X, columns=[f'feature_{i+1}' for i in range(X.shape[1])])
data_df['target'] = y

data_df
```

```python
# Scatter plot of two informative features, colored by class
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k', s=50, alpha=0.7)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Scatter Plot of Synthetic Classification Data')
plt.show()
```
