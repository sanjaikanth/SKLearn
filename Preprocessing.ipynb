```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pandas as pd

# Sample data
data = np.array([[1, 200], [2, 300], [3, 400], [4, 500], [5, 600]])
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

```

```python
# Apply MinMaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaled_data = minmax_scaler.fit_transform(df)

```

```python
df
```

```python
print("\nMin-Max Scaled Data:\n") 
pd.DataFrame(minmax_scaled_data, columns=['Feature1', 'Feature2'])
```

```python
from sklearn.preprocessing import Normalizer

# Apply Normalizer
normalizer = Normalizer()
normalized_data = normalizer.fit_transform(df)

print("\nNormalized Data:\n", pd.DataFrame(normalized_data, columns=['Feature1', 'Feature2']))
```

```python
import pandas as pd

# Sample data
data = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red'],
    'size': ['S', 'M', 'L', 'M', 'L'],
    'label': ['cat', 'dog', 'cat', 'dog', 'cat']
})
data
```

```python
from sklearn.preprocessing import OneHotEncoder

# Apply OneHotEncoder
onehot_encoder = OneHotEncoder(sparse=False)  # Setting sparse=False to get a dense matrix for readability
encoded_data = onehot_encoder.fit_transform(data[['color', 'size']])

# Convert the output to a DataFrame for better readability
encoded_df = pd.DataFrame(encoded_data, columns=onehot_encoder.get_feature_names_out(['color', 'size']))
encoded_df
```

```python
import pandas as pd
import numpy as np

# Sample data with missing values
data = pd.DataFrame({
    'age': [25, np.nan, 30, 35, np.nan],
    'income': [50000, 60000, np.nan, 45000, 52000],
    'education': ['Bachelors', 'Masters', 'PhD', np.nan, 'Masters']
})
print("Original Data with Missing Values:\n") 
data
```

```python
from sklearn.impute import SimpleImputer

# Impute numerical columns with the mean
num_imputer = SimpleImputer(strategy='mean')
data[['age', 'income']] = num_imputer.fit_transform(data[['age', 'income']])

# Impute categorical column with the most frequent value
cat_imputer = SimpleImputer(strategy='most_frequent')
data['education'] = cat_imputer.fit_transform(data[['education']])

print("\nSimpleImputer Result:\n")
data
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Sample text data
documents = [
    "I love machine learning and artificial intelligence",
    "Machine learning is a fascinating field",
    "I am learning how to use TfidfVectorizer",
    "Artificial intelligence is transforming industries",
    "Learning machine learning concepts is fun!"
]
# Initialize the TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the data
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Convert the output to a DataFrame for readability
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Display the TF-IDF matrix
print("TF-IDF Matrix:\n")
tfidf_df
```
