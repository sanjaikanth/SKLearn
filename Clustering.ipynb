```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Generate synthetic data for clustering
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# Initialize and train the KMeans model
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)
centers = kmeans.cluster_centers_

# Plot original data and KMeans results side by side
fig, axs = plt.subplots(1, 2, figsize=(15, 6))

# Plot original data
axs[0].scatter(X[:, 0], X[:, 1], s=50, color="blue")
axs[0].set_title("Original Data for Clustering")

# Plot KMeans clustering results
axs[1].scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
axs[1].scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X', label="Centers")
axs[1].set_title("KMeans Clustering")
axs[1].legend()

plt.tight_layout()
plt.show()
```
